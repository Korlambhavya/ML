```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=pd.read_csv("wine-red.csv")
data.corr().abs().nlargest(3,'quality').index #taking high correlated features
x=data[['alcohol','volatile acidity']] #multiple variables
y=data['quality'] #target feature
from sklearn.model_selection import train_test_split
#splitting x and y into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,
                                               random_state=2019)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)  #fit x and y
y_pred=model.predict(x_test) #predicted values
print(model.intercept_) # intercept
print(model.coef_) # coefficients of both alcohol & volatile acidity
print(model.score(x_test,y_test))
#plt.subplot(121) # 1row,2col,chart 1
# plt.plot(y_test,y_pred,'r')
#plt.xlabel("Alcohol")
#plt.ylabel("Quality")
#plt.subplot(122)  # 1 row & 2 col & chart 2
#plt.plot(y_test,y_predict,'y^')
#plt.xlabel("Volatile-acidity")
#plt.ylabel("Quality")
plt.subplot(1,2,1)
plt.scatter(x_test['alcohol'], y_pred)
plt.subplot(1,2,2)
plt.scatter(x_test['volatile acidity'], y_pred)

```
